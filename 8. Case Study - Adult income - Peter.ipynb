### **19. Who has a better chance to get salary >50k, male or female? **

```python
count_male_high_salary = data2[(data2['gender'] == 'Male') & (data2['Salary'] == '>50k')].shape[0]
count_female_high_salary = data2[(data2['gender'] == 'Female') & (data2['Salary'] =='>50k' )].shape[0]
total_male = data2[data2['gender'] == 'Male'].shape[0]
total_female = data2[data2['gender'] == 'Female'].shape[0]

# Calculate the percentage of males and females earning >50000
percentage_male_high_salary = (count_male_high_salary / total_male) * 100
percentage_female_high_salary = (count_female_high_salary / total_female) * 100

print(f"Percentage of males earning >50000: {percentage_male_high_salary:.2f}%")
print(f"Percentage of females earning >50000: {percentage_female_high_salary:.2f}%")
```
